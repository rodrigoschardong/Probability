```python
#Coin Probability

import numpy as np

#Libs Just to Display
from matplotlib import pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

from A_coin_probability import Probability_A
pf = Probability_A()

def experiment_plot(probability, chances):
    results = pf.Experiment( probability, chances)
    f, ax = plt.subplots(ncols = 3, figsize = (20,5))

    ax[0].bar(np.arange(len(results)),results)
    ax[0].set_ylim(bottom = 0.0,top = 1.0)
    ax[0].set_xlabel("Number of Toss")
    ax[0].set_ylabel("Results")
    ax[0].set_title("Resume of Toss")

    bars = [(1 - pf.Frequency(results)) * 100, pf.Frequency(results) * 100] 
    ax[1].bar([0,1],bars)
    ax[1].set_ylim(bottom = 0.0,top = 100.0)
    ax[1].set_xlabel("False / True")
    ax[1].set_ylabel("Frequency")
    ax[1].set_title("Result of Toss")

    historic = pf.Frequencies(probability, chances)
    ax[2].plot(historic)
    ax[2].axhline(y = probability, color ='r', linestyle = '--')
    ax[2].set_ylim(bottom = 0.0,top = 1.0)
    ax[2].set_xlabel("Number of Toss")
    ax[2].set_ylabel("Frequencies")
    ax[2].set_title("Resume of Frequencies")
    plt.show()

interact(experiment_plot,
        probability = widgets.FloatSlider(value = 0.5, min = 0, max = 1, step = 0.01),
        chances = widgets.IntSlider(value = 5, min = 2, max = 1000, step = 1)
        )
```

```python
#Lottery

import numpy as np

#Libs Just to Display
from matplotlib import pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import math
from scipy.special import factorial

from C_lottery_probability import Probability_C

def experiment_plot():
    sampleUniverse = 60
    luckNumbers = 6
    lottery = Probability_C(sampleUniverse, luckNumbers)

    number_of_attemps = 200
    purchase_six = lottery.Plays(6, number_of_attemps)
    purchase_seven = lottery.Plays(7, number_of_attemps)
    purchase_eight = lottery.Plays(8, number_of_attemps)

    f, ax = plt.subplots(ncols = 3, figsize = (20,5))

    ax[0].plot(purchase_six, color='blue')
    #ax[0].set_ylim(bottom = 0.0,top = 0.00001)
    ax[0].set_xlabel("Numero de Cotas")
    ax[0].set_ylabel("Probabilidade de ganho")
    ax[0].set_title("Aposta em 6 nÃºmeros")

    ax[1].plot(purchase_seven, color='red')
    #ax[1].set_ylim(bottom = 0.0,top = 0.00001)
    ax[1].set_xlabel("Numero de Cotas")
    ax[1].set_ylabel("Probabilidade de ganho")
    ax[1].set_title("Aposta em 7 nÃºmeros")

    ax[2].plot(purchase_eight, color='green')
    #ax[2].set_ylim(bottom = 0.0,top = 0.00001)
    ax[2].set_xlabel("Numero de Cotas")
    ax[2].set_ylabel("Probabilidade de ganho")
    ax[2].set_title("Aposta em 8 nÃºmeros")
    plt.show()

    # Plotting all data in a single graph
    plt.figure(figsize=(10, 6))
    plt.plot(purchase_six, label='6 nÃºmeros', color='blue')
    plt.plot(purchase_seven, label='7 nÃºmeros', color='red')
    plt.plot(purchase_eight, label='8 nÃºmeros', color='green')
    plt.xlabel("Numero de Cotas")
    plt.ylabel("Probabilidade de ganho")
    plt.title("Probabilidade de ganho com diferentes apostas")
    plt.legend()  # Show legend with labels
    plt.show()

    #Chance de ganhar por real gasto na mega da virada

    cost_six_numbers = 5
    cost_seven_numbers = 35
    cost_eight_numbers = 140

    chances_per_real_6_numbers = []
    chances_per_real_7_numbers = []
    chances_per_real_8_numbers = []

    for i in range(number_of_attemps):
        chances_per_real_6_numbers.append(purchase_six[i] / cost_six_numbers)
        chances_per_real_7_numbers.append(purchase_seven[i] / cost_seven_numbers)
        chances_per_real_8_numbers.append(purchase_eight[i] / cost_eight_numbers)

    f, bx = plt.subplots(ncols = 3, figsize = (20,5))

    bx[0].plot(chances_per_real_6_numbers, color='blue')
    #ax[0].set_ylim(bottom = 0.0,top = 0.00001)
    bx[0].set_xlabel("Numero de Cotas")
    bx[0].set_ylabel("Probabilidade de ganho por real")
    bx[0].set_title("Aposta em 6 nÃºmeros")

    bx[1].plot(chances_per_real_7_numbers, color='red')
    #ax[1].set_ylim(bottom = 0.0,top = 0.00001)
    bx[1].set_xlabel("Numero de Cotas")
    bx[1].set_ylabel("Probabilidade de ganho por real")
    bx[1].set_title("Aposta em 7 nÃºmeros")

    bx[2].plot(chances_per_real_8_numbers, color='green')
    #ax[2].set_ylim(bottom = 0.0,top = 0.00001)
    bx[2].set_xlabel("Numero de Cotas")
    bx[2].set_ylabel("Probabilidade de ganho por real")
    bx[2].set_title("Aposta em 8 nÃºmeros")
    plt.show()

    # Plotting all data in a single graph
    plt.figure(figsize=(10, 6))
    plt.plot(chances_per_real_6_numbers, label='6 nÃºmeros', color='blue')
    plt.plot(chances_per_real_7_numbers, label='7 nÃºmeros', color='red')
    plt.plot(chances_per_real_8_numbers, label='8 nÃºmeros', color='green')
    plt.xlabel("Numero de Cotas")
    plt.ylabel("Probabilidade de ganho por real")
    plt.title("Probabilidade de ganho por real com diferentes apostas")
    plt.legend()  # Show legend with labels
    plt.show()


experiment_plot()
```
